Treinamento Inicial do Modelo

```python
# ==============================================================================
# ğŸš€ TREINAMENTO TURBO YOLOv8 - DETECTOR DE CAMPOS (NOVO NOTEBOOK)
# ==============================================================================
# 1. InstalaÃ§Ã£o e Setup
# ==============================================================================
print("â³ Instalando bibliotecas...")
!pip install ultralytics -q

import os
import shutil
import yaml
from google.colab import drive
from ultralytics import YOLO

# Montar o Google Drive
if not os.path.exists('/content/drive'):
    print("ğŸ“‚ Montando Google Drive...")
    drive.mount('/content/drive')

# ==============================================================================
# 2. ConfiguraÃ§Ãµes de Caminhos (Ajuste se necessÃ¡rio)
# ==============================================================================
# Onde estÃ¡ o seu ZIP no Drive
ZIP_NO_DRIVE = '/content/drive/MyDrive/meus_modelos_ia/datasets/dataset_completo.zip'

# Onde vamos descompactar (SSD Local do Colab - Super RÃ¡pido)
PASTA_LOCAL_COLAB = '/content/dataset_rapido'

# Onde vamos salvar o modelo treinado (No Drive, para nÃ£o perder)
PASTA_SAIDA_MODELO = '/content/drive/MyDrive/document-classifier/models/field_detector'

# ==============================================================================
# 3. PreparaÃ§Ã£o dos Dados (CÃ³pia e Unzip)
# ==============================================================================
if not os.path.exists(PASTA_LOCAL_COLAB):
    print(f"ğŸš€ Copiando ZIP do Drive para o Colab (Isso acelera o treino em 100x)...")
    shutil.copy(ZIP_NO_DRIVE, '/content/temp.zip')

    print(f"ğŸ“¦ Descompactando... (Pode levar 1 minutinho)")
    # O comando -q Ã© para nÃ£o poluir a tela com milhares de nomes de arquivos
    !unzip -q /content/temp.zip -d {PASTA_LOCAL_COLAB}
    print("âœ… Dataset descompactado e pronto para uso!")
else:
    print("âœ… Dataset jÃ¡ estava descompactado. Pulando etapa.")

# ==============================================================================
# 4. Criar Arquivo de ConfiguraÃ§Ã£o (data.yaml) Local
# ==============================================================================
yaml_content = {
    'path': PASTA_LOCAL_COLAB,  # Aponta para a pasta local rÃ¡pida
    'train': 'images/train',
    'val': 'images/val',
    'names': {
        0: 'cpf',
        1: 'nome',
        2: 'data_nasc',
        3: 'rg_numero',
        4: 'filiacao',
        5: 'categoria',
        6: 'validade',
        7: 'data_exp',
        8: 'naturalidade'
    }
}

LOCAL_YAML = '/content/data_local.yaml'
with open(LOCAL_YAML, 'w') as f:
    yaml.dump(yaml_content, f)

print(f"ğŸ“ Arquivo de configuraÃ§Ã£o criado em: {LOCAL_YAML}")

# ==============================================================================
# 5. Treinamento Blindado (Com Checkpoints e ProteÃ§Ã£o)
# ==============================================================================
def treinar():
    print("\nğŸ”¥ INICIANDO TREINAMENTO TURBO...")
    print(f"   ğŸ“‚ Salvando resultados em: {PASTA_SAIDA_MODELO}")
    print("   ğŸ›‘ Para parar manualmente, clique no botÃ£o de STOP (o YOLO salva tudo antes de sair).")

    # Carregar modelo YOLO leve
    model = YOLO('yolov8n.pt')

    try:
        model.train(
            data=LOCAL_YAML,
            epochs=100,
            imgsz=640,
            batch=16,
            project=PASTA_SAIDA_MODELO,
            name='treino_final_v1',

            # --- ConfiguraÃ§Ãµes de Salvamento ---
            save=True,            # Salva checkpoints
            save_period=10,       # Salva a cada 10 Ã©pocas (epoch10.pt, epoch20.pt...)
            patience=15,          # Se ficar 15 Ã©pocas sem melhorar, ele para sozinho (Early Stopping)
            exist_ok=True,        # Continua na mesma pasta se rodar de novo
            verbose=True
        )
        print("\nğŸ† TREINO FINALIZADO COM SUCESSO!")

    except KeyboardInterrupt:
        print("\n\nğŸ›‘ INTERRUPÃ‡ÃƒO MANUAL DETECTADA!")
        print("   O treinamento foi salvo.")
        print("   VocÃª pode retomar depois usando o arquivo 'last.pt'.")

    print(f"\nğŸ“‚ Seus modelos finais estÃ£o em: {PASTA_SAIDA_MODELO}/treino_final_v1/weights/")
    print("   -> best.pt (Use este para produÃ§Ã£o)")
    print("   -> last.pt (Ãšltimo estado)")

# Rodar!
treinar()
```

Fine tuning do modelo com parametros modificados, para manter a precisÃ£o e melhor as mÃ©tricas de loss e map 50-95

```python
# ==============================================================================
# ğŸš€ TREINAMENTO TURBO YOLOv8 - DETECTOR DE CAMPOS (NOVO NOTEBOOK)
# ==============================================================================
# 1. InstalaÃ§Ã£o e Setup
# ==============================================================================
print("â³ Instalando bibliotecas...")
!pip install ultralytics -q

import os
import shutil
import yaml
from google.colab import drive
from ultralytics import YOLO

# Montar o Google Drive
if not os.path.exists('/content/drive'):
    print("ğŸ“‚ Montando Google Drive...")
    drive.mount('/content/drive')

# ==============================================================================
# 2. ConfiguraÃ§Ãµes de Caminhos (Ajuste se necessÃ¡rio)
# ==============================================================================
# Onde estÃ¡ o seu ZIP no Drive
ZIP_NO_DRIVE = '/content/drive/MyDrive/meus_modelos_ia/datasets/dataset_completo.zip'

# Onde vamos descompactar (SSD Local do Colab - Super RÃ¡pido)
PASTA_LOCAL_COLAB = '/content/dataset_rapido'

# Onde vamos salvar o modelo treinado (No Drive, para nÃ£o perder)
PASTA_SAIDA_MODELO = '/content/drive/MyDrive/document-classifier/models/field_detector'

# ==============================================================================
# 3. PreparaÃ§Ã£o dos Dados (CÃ³pia e Unzip)
# ==============================================================================
if not os.path.exists(PASTA_LOCAL_COLAB):
    print(f"ğŸš€ Copiando ZIP do Drive para o Colab (Isso acelera o treino em 100x)...")
    shutil.copy(ZIP_NO_DRIVE, '/content/temp.zip')

    print(f"ğŸ“¦ Descompactando... (Pode levar 1 minutinho)")
    # O comando -q Ã© para nÃ£o poluir a tela com milhares de nomes de arquivos
    !unzip -q /content/temp.zip -d {PASTA_LOCAL_COLAB}
    print("âœ… Dataset descompactado e pronto para uso!")
else:
    print("âœ… Dataset jÃ¡ estava descompactado. Pulando etapa.")

# ==============================================================================
# 4. Criar Arquivo de ConfiguraÃ§Ã£o (data.yaml) Local
# ==============================================================================
yaml_content = {
    'path': PASTA_LOCAL_COLAB,  # Aponta para a pasta local rÃ¡pida
    'train': 'images/train',
    'val': 'images/val',
    'names': {
        0: 'cpf',
        1: 'nome',
        2: 'data_nasc',
        3: 'rg_numero',
        4: 'filiacao',
        5: 'categoria',
        6: 'validade',
        7: 'data_exp',
        8: 'naturalidade'
    }
}

LOCAL_YAML = '/content/data_local.yaml'
with open(LOCAL_YAML, 'w') as f:
    yaml.dump(yaml_content, f)

print(f"ğŸ“ Arquivo de configuraÃ§Ã£o criado em: {LOCAL_YAML}")

# ==============================================================================
# 5. Treinamento Blindado (Com Checkpoints e ProteÃ§Ã£o)
# ==============================================================================
def treinar():
    print("\nğŸ”¥ INICIANDO TREINAMENTO (FINE-TUNING CONTROLADO)...")

    # âš ï¸ IMPORTANTE: Carregue o modelo do treino incial!
    caminho_modelo_bom = '/content/drive/MyDrive/document-classifier/models/field_detector/treino_final_v1/weights/best.pt'

    if not os.path.exists(caminho_modelo_bom):
        print(f"âŒ ERRO CRÃTICO: NÃ£o achei o modelo bom em: {caminho_modelo_bom}")
        return

    print(f"   ğŸ§  Carregando a inteligÃªncia do modelo de ouro: {caminho_modelo_bom}")
    model = YOLO(caminho_modelo_bom)

    try:
        model.train(
            data=LOCAL_YAML,
            epochs=30,            # 30 Ã©pocas deve ser suficiente para fine-tuning mas pode ser aumentado ou diminuido conforme a ncessidade
            imgsz=640,
            batch=16,
            optimizer='AdamW',
            lr0=0.0001,
            lrf=0.01,
            warmup_epochs=0,
            amp=False,
            # --- ConfiguraÃ§Ãµes de Salvamento ---
            project=PASTA_SAIDA_MODELO,
            name='treino_refinado_v3',
            save=True,
            patience=10,
            exist_ok=True,
            verbose=True
        )
        print("\nğŸ† TREINO FINALIZADO COM SUCESSO!")

    except KeyboardInterrupt:
        print("\n\nğŸ›‘ INTERRUPÃ‡ÃƒO MANUAL DETECTADA!")

    print(f"\nğŸ“‚ Seus modelos finais estÃ£o em: {PASTA_SAIDA_MODELO}/treino_refinado_v3/weights/")


treinar()
```
